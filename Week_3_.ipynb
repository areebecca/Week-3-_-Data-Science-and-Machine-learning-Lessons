Lambda Functions:

Lambda Functions, also known as anonymous functions, are small, single expressions that can be defined in line using the lambda key word

They are useful for creating quick, throwaway functions without the need for a formal def statement

````python
# Adding the numbers
add =  lambda x,y: x+y
print(add(2,6))
````

````python
# Getting the product of numbers
product = lambda x,y: x*y
print(product(4,6))
````

````python
students =[
    {'name': 'Alice', 'grade': 85},
    {'name': 'Bob', 'grade': 92},
]
sorted_students = sorted(students, key=lambda student: student['grade'],reverse=True)
print(sorted_students)
````

Built in Functions

Python comes with several built in functions that perform common operations efficiently

map(function,iterable):

````python
# Finding the square of the numbers in the list
numbers=[1,2,3,4,5]
squared= map(lambda x: x**2, numbers)
print(list(squared))
````

````python
# coverting temperatures from celsius to Fahrenheit using map
temps_celsius= [0,10,20,30]
temps_fahrenheit = map(lambda c: (c*9/5)+32, temps_celsius)
print(list(temps_fahrenheit))
````

````python
# Capitalizing the first letter of each word in alist of strings using map
words=['hello','world']
capitilized_words = map(lambda w: w.capitalize(), words)
print(list(capitilized_words))
````

filter(function, iterable): Filters elements from an iterable based on the given function's truthiness

````python
# finding the even numbers in the list
numbers= [1,2,3,4,5,6]
evens =  filter(lambda x: x%2 ==0, numbers)
print(list(evens))
````

````python
# finding negative numbers in a list
number = [1, -2,3,-6,4]
negative_numbers= filter(lambda x: x<0, number)
print(list(negative_numbers))
````

````python
# finding positive numbers in a list
number = [1, -2,3,-6,4]
positive_numbers= filter(lambda x: x>0, number)
print(list(positive_numbers))
````

reduce(function,iterable[,initializer]):

Applies the given function cumulatively to the items of the iterable, from left to right , to reduce the iterable to a single value

````python
# Finding the product of the elements in the list
from functools import reduce
numbers =[1,2,3,4,5]
product = reduce(lambda x,y: x*y, numbers)
print(product)
````

````python
#finding the maximum element in alist using reduce
numbers = [1,2,3,4,5]
max_num = reduce(lambda x,y: x if x>y else y, numbers)
print(max_num)
````

````python
#finding the maximum element in alist using reduce
numbers= [29,2,3,40,87]
max_num= reduce(lambda x,y: x if x>y else y, numbers)
print(max_num)
````

sum(iterable[,start]):

Returns the sum of all elements in the iteration

````python
# finding the sum of the numbers in the list
numbers= [1,2,3,4]
total = sum(numbers)
print(total)
````

max(iterable[,key,default])

Returns the maximum element from the iteration

````python
#finding the maximum value in a lis
numbers= [2,4,6,7,50,9]
max_num= max(numbers)
print(max_num)
````

````python
# finding the longest string in alist
strings= ['apple','banana','cherry']
longest_string= max(strings, key=len)
print(longest_string)
````

min(iterable[,key,default])

returns the minimum element from the iterable

````python
#finding the minimum element from iterable
numbers= [1,2,3,4,5]
min_num= min(numbers)
print(min_num)
````

````python
# finding the minimum value in a dictionary
scores= {'Rebecca':90,'Mercy':76}
min_score= min(scores.values())
print(min_score)
````

Data Combination

Is the process of joining, merging or concatenating multiple pandas data structure in to a single data structure.
The most common ways of combining Dataframes are concatenation, joining and merge

Concatenation

```
This is formatted as code: using concat function to combine two or more dataframes with the same columns either vertically or horizontally
```



if they don't have the same column, we can't concatenate

````python
df_1= pd.DataFrame({'A':[2,4,6],'B':['x','y','z']})
df_2= pd.DataFrame({'A':[20,20,5],'B':['w','v','u']})
````

````python
df_1
````

````python
df_2
````

````python
# vertical concatenation (row wise), axis=0 represents vertical concatenation
vertical_com = pd.concat([df_1,df_2], axis=0, keys=['df_1','df_2'])
vertical_com
````

````python
# Vertical concatenation (row-wise)
vertical_com = pd.concat([df_1, df_2], axis=0, ignore_index=True)
vertical_com
````

````python
# Horizontal concatenation (column-wise)
horizontal_com = pd.concat([df_1, df_2], axis=1, keys=["df_1","df_2"])
horizontal_com
````
